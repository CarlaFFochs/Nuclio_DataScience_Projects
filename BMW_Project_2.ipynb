```python
# Vamos a checkear las variables booleanas
df["volante_regulable"].describe()
```

```python
df["volante_regulable"] = df["volante_regulable"].astype(bool)
```

```python
df["volante_regulable"].dtype
```

```python
lista_booleanas = ["volante_regulable", "aire_acondicionado", "camara_trasera", "asientos_traseros_plegables", "elevalunas_electrico", "bluetooth", "alerta_lim_velocidad"]
```

```python
for item in lista_booleanas:
    if item == item:
        df[item] = df[item].astype(bool)
```

```python
df.info()
```

Ahora vamos a imputar nulos a "mano"

```python
df.isnull().sum().sum()
```

Empezamos por las numericas

```python
df.select_dtypes(include = ['number']).isnull().sum()
```

```python
list_numerical = df.select_dtypes(include='number')
list_numerical
```

```python
for item in list_numerical:
   if item == item:
      df[item].fillna(df[item].mean, inplace = True)
```

Nos cargamos los decimales, ya que todos son 0 y no nos aportan ningún valor (al contrario, puede que en un futuro nos den problemas)
